```python
materias = ['matematica', 'historia', 'biologia']
notas = (8.5, 7.0, 9.2)
info_aluno = {'nome':'Geovana', 'idade':26,'curso':'Engenharia automobilística' }
print(f'{info_aluno}, {notas}, {materias}')
```

```python
acessorrios = ['tapetes','câmbio','volante']
especificacoes = ('136cv', '42l')
carro = {'marca':'byd', 'modelo':'king', 'ano':'2025'}
print(f'{acessorrios} {especificacoes} {carro}')
```

```python
categorias = ['Ribs' ,'The Outbacker', 'smores']
especificacoes = ('61.58', '85.54', '45')
produto = {'nome': 'ribs', 'marca': 'Outback', 'ano':'2026'}
print(categorias)
print(especificacoes)
print(produto)
```
